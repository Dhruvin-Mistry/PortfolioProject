```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pyplot import figure
import seaborn as sns
matplotlib.rcParams['figure.figsize']= (12,8)
```

```python
df = pd.read_csv(r"C:\Users\Dhruvin\Desktop\Portfolio Project\Movies Project\movies.csv")
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['budget'].count()
```

```python
# Dropping Null Values
df = df.dropna()
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df.head()
```

```python
df.drop(columns = 'released',inplace=True)
```

```python
df.head()
```

```python
df.sort_values(by = ['gross'],inplace=False, ascending = False)
```

```python
#TO see all the data
pd.set_option('display.max_rows',None)
df.tail()
```

```python
#Dropping Duplicates

df.drop_duplicates(inplace = True)
```

```python
df.head()
```

```python
#Scatter Plot of Budget vs Revenue

plt.scatter(x= df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
```

```python
#Use of Seaborn

sns.regplot(x= 'budget',y='gross',data =df,scatter_kws = {"color":"red"},line_kws={"color":"blue"})
```

```python
df.corr() #High Correlation between budget and gross (Pearson Correlation)
```

```python
correlation_matrix = df.corr() 
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
```

```python
#Correlation of Non Numeric Features

df_numeric = df

for col_name in df_numeric.columns:
    if(df_numeric[col_name].dtype == 'object'):
        df_numeric[col_name] = df_numeric[col_name].astype('category')
        df_numeric[col_name] = df_numeric[col_name].cat.codes
df_numeric
```

```python
correlation_matrix = df_numeric.corr() 
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
```

```python
df_numeric.corr()
```

```python
c1 = df_numeric.corr()
corr_pairs = c1.unstack()
corr_pairs
```

```python
corr_sorted = corr_pairs.sort_values()
corr_sorted
```

```python
high_corr = corr_sorted[(corr_sorted)>0.5 ]
high_corr


#Conclusion: High correlation between Bugdget and Gross Earnings 
#Also High Correlation between Votes and Gross Earnings
```
